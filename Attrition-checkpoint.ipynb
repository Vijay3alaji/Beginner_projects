```python
import pandas as pd
import numpy as np
```

```python
train = pd.read_csv(r"D:\PGP DS\ML\PROJECT\classification\train_HR.csv")
test = pd.read_csv(r"D:\PGP DS\ML\PROJECT\classification\test_HR.csv")
```

```python
test.shape
```

```python
train["education"].fillna(train["education"].mode()[0],inplace = True)
train["previous_year_rating"].fillna(train["previous_year_rating"].mean(),inplace = True)
```

```python
test["previous_year_rating"].fillna(test["previous_year_rating"].mean(),inplace = True)
test["education"].fillna(test["education"].mode()[0],inplace = True)
```

```python
x_train = train.iloc[:,1:13]
x_test = test.iloc[:,1:13]
y_train = train.iloc[:,13].values
```

```python
dataset = pd.concat([x_train,x_test],axis=0)
```

```python
dataset
```

```python
dummies = pd.get_dummies(data=dataset, columns=["department","region","education","gender","recruitment_channel"],drop_first = True)
```

```python
dummies
```

```python
dummies.info(object)
```

```python
dummies["intercept"] = 1
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
dummies.drop(columns = ['avg_training_score',"department_Operations",'department_Procurement','department_Sales & Marketing',"region_region_11",
        'region_region_13','region_region_15','region_region_16','region_region_2','region_region_22','region_region_23',
        'region_region_26','region_region_27','region_region_28','region_region_31','region_region_4','region_region_7','intercept'],inplace = True)
```

```python
dummies.drop(columns = ['age','gender_m','length_of_service','previous_year_rating','no_of_trainings'],inplace = True)
```

```python
dummies
```

```python
ds_train=dummies.iloc[:54808,:]
ds_test=dummies.iloc[:23490,:]
```

## standard scalar

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
```

```python
x_train=sc.fit_transform(ds_train)
x_test=sc.transform(ds_test)
```

## Hypertuning

```python
from sklearn.model_selection import GridSearchCV
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
classifier = RandomForestClassifier(random_state=0)
```

```python
param_gride={'n_estimators':[10,20,25,30,75,100],
            'criterion':['entropy','gini'],
            'bootstrap':[True, False]}
```

```python
gr=GridSearchCV(estimator = classifier, param_grid= param_gride, cv=5, n_jobs=-1)
```

```python
gr.fit(x_train,y_train)
```

```python
gr.best_params_
```

```python
gr.best_score_
```
